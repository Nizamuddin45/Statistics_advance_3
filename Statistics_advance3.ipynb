**Question 1: What is hypothesis testing in statistics?**

Answer:- Hypothesis testing in statistics is a formal method used to make decisions or inferences about population parameters based on sample data. It helps determine whether there is enough evidence to support a specific claim or hypothesis about a population.

**Key Components of Hypothesis Testing:**

# 1. Null Hypothesis:
The default or initial assumption. It usually states that there is no effect or no difference.
Example : The average height of students is 170 cm.

# 2. Alternative Hypothesis:

The claim you want to test. It represents an effect, difference, or relationship.
Example: The average height of students is not 170 cm.

# 3. Significance Level:
The probability of rejecting the null hypothesis when it is actually true. Common choices are 0.05, 0.01.

# 4. Test Statistic:
A value calculated from the sample data that is used to test the hypothesis.

# 5. P-value:
The probability of observing the sample data (or something more extreme) if the null hypothesis is true.

**Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.**
**Answer:-**
| Feature                               | **Z-test**                           | **T-test**                                  |
| ------------------------------------- | ------------------------------------ | ------------------------------------------- |
| **Population Standard Deviation (Ïƒ)** | **Known**                            | **Unknown**                                 |
| **Sample Size**                       | **Large** (n â‰¥ 30)                   | **Small** (n < 30)                          |

| **Tail Shape**                        | Narrower tails                       | Wider tails (accounts for more variability) |
| **Use case**                          | Testing population mean with known Ïƒ | Testing mean when Ïƒ is unknown              |


**When to use Z-Test:-**
1. The sample size is large (n â‰¥ 30)

2. The population standard deviation (Ïƒ) is known

3. Data is normally distributed or sample is large enough for the Central Limit Theorem to apply

**When to use T-Test:-**

1. The population standard deviation is unknown.

2. The sample size is small (n â‰¤ 30).

3. The data is approximately normally distributed.


**Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.**

**Answer:-** Hereâ€™s a Python program to generate a binomial distribution with parameters:

n = 10 (number of trials)

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10       # number of trials
p = 0.5      # probability of success
size = 1000  # number of samples

# Generate binomial distribution data
data = np.random.binomial(n=n, p=p, size=size)

# Plot the histogram
plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]**

**Answer:-**

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Step 1: Sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)
sigma = 0.5  # Assumed known population standard deviation
mu_0 = 50    # Population mean under null hypothesis

# Step 2: Z-statistic calculation
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Step 3: P-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 4: Output
print(f"Sample Mean = {sample_mean:.4f}")
print(f"Z-statistic = {z_stat:.4f}")
print(f"P-value = {p_value:.4f}")

# Step 5: Interpret result at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

**Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Simulate data from a normal distribution
np.random.seed(42)  # For reproducibility
mu, sigma = 50, 10  # Mean and standard deviation
sample_size = 100
data = np.random.normal(mu, sigma, sample_size)

# 2. Calculate 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
conf_level = 0.95
alpha = 1 - conf_level

# Using the t-distribution for the confidence interval
t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1)
margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))
conf_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: {conf_interval[0]:.2f} to {conf_interval[1]:.2f}")

# 3. Plot the data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Mean = {sample_mean:.2f}')
plt.axvline(conf_interval[0], color='green', linestyle='--', label=f'95% CI Lower = {conf_interval[0]:.2f}')
plt.axvline(conf_interval[1], color='green', linestyle='--', label=f'95% CI Upper = {conf_interval[1]:.2f}')
plt.title('Histogram of Simulated Normal Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Question 9: Write a Python function to calculate the Z- scores from the dataset and visualize the standardized data using a histogram. Explain what the Z- scores represent in terms of standard deviations from the mean.**

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset.

    Parameters:
        data (array-like): The original dataset.

    Returns:
        z_scores (np.ndarray): Standardized values (Z-scores).
    """
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

# Generate sample data
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=1000)

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Print first few Z-scores
print("First 5 Z-scores:")
print(z_scores[:5])

# Plot histogram of Z-scores
plt.figure(figsize=(10, 6))
plt.hist(z_scores, bins=30, color='purple', edgecolor='black', alpha=0.7)
plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
plt.title('Histogram of Z-scores (Standardized Data)')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
